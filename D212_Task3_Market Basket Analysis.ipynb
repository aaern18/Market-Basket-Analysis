### Prepare the Dataset

In the code below the dropna() is used to remove rows with NaN values. The axis=0 parameter specifies that rows should be dropped, and how='all' indicates that a row will only be dropped if all its values are NaN.

```python
# remove the rows which has all null values 
df_telco = df_telco.dropna(axis=0, how='all')
```

### Convert Data to Transaction Format

Convert the dataset into a transaction format that is suitable for Market Basket Analysis. This involves creating a list of lists, where each inner list represents a transaction and contains the items present in that transaction.



```python
# Create a list of list for each row in DataFrame
transactions = []
for i, transaction in df_telco.iterrows():
    transactions.append(transaction[transaction.notna()].tolist())
```

```python
print ("Transaction 1:" , transactions[0])
print ('************************************************************************')
print ("Transaction 5:" , transactions[4])
print ('************************************************************************')
print ("Transaction 30:" , transactions[29])
```

### Encode the Transaction Data

Encode the transaction data using the TransactionEncoder class from mlxtend.preprocessing. This will transform the transaction data into a one-hot encoded format.

```python
# Instantiate transaction encoder 
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

# convert into DataFrame 
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

### Market Basket Analysis 

### Generate Frequent Itemsets

Use the apriori function from mlxtend.frequent_patterns to generate frequent itemsets from the encoded data.

```python
frequent_itemsets = apriori(df_encoded, min_support=0.05, use_colnames=True)
```

```python
frequent_itemsets
```

```python
# print number of items
print("There are",frequent_itemsets.shape[0], "sets of items." )
```

### Generate Association Rules

Used the association_rules function from mlxtend.frequent_patterns to generate association rules from the frequent itemsets.

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
```

```python
rules
```

```python
# Iterate over each rule and display the relevant information
for idx, rule in rules.iterrows():
    antecedents = ' + '.join(list(rule['antecedents']))
    consequents = ' + '.join(list(rule['consequents']))
    confidence = rule['confidence']
    support = rule['support']
    lift = rule['lift']
    
    print(f"Rule #{idx+1}: {antecedents} -> {consequents}")
    print(f"Confidence: {confidence:.2f}")
    print(f"Support: {support:.2f}")
    print(f"Lift: {lift:.2f}")
    print()
```

###  Top Three Rules 

The support matrix indicates the proportion of transactions in the dataset that contains a specific itemset. 

Higher support values indicate that the itemsets are popular and frequently co-purchased by customers. 

```python
# Sort the rules based on support in descending order
rules_sorted_by_support = rules.sort_values(by='support', ascending=False)
```

```python
# Select the top three rules with the highest lift
top_three_rules_support = rules_sorted_by_support.head(3)
```

```python
top_three_rules_support 
```

```python
# Iterate over each rule in the DataFrame
for index, rule in top_three_rules_support.iterrows():
    antecedents = ', '.join(list(rule['antecedents']))
    consequents = ', '.join(list(rule['consequents']))
    confidence = rule['confidence']
    support = rule['support']
    lift = rule['lift']

    # Print the rule and its corresponding metrics
    print(f"Rule: {antecedents} -> {consequents}")
    print(f"Confidence: {confidence:.2f}")
    print(f"Support: {support:.2f}")
    print(f"Lift: {lift:.2f}")
    print()
```

**Rule 1** : 

Customers who purchased the "Dust-Off Compressed Gas 2 pack" have a 25% probability of also purchasing the "VIVO Dual LCD Monitor Desk mount". 


**Rule 2** : 

Customers who purchased the "VIVO Dual LCD Monitor Desk mount" have a 34% probability of also purchasing the "Dust-Off Compressed Gas 2 pack"


**Rule 3** : 

Customers who purchased the "HP 61 ink" have a 32% probability of also purchasing the "Dust-Off Compressed Gas 2 pack".
